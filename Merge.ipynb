 ## ESG + 최종

```python
ESG = pd.read_csv('./datasets/insight_data/ESG최종.csv',encoding='utf-8-sig',index_col=0)
insight = pd.read_csv('./datasets/insight_data/끝.csv',encoding='utf-8-sig',index_col=0)
```
